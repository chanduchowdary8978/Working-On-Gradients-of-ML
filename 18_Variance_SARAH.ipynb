```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

```python
def generate_data(n=1000, d=10):
    X = np.random.randn(n, d)
    true_w = np.random.randn(d)
    logits = X @ true_w
    y = (logits > 0).astype(int)
    return X, y

X, y = generate_data()
n, d = X.shape
```

```python
def sigmoid(z):
    z = np.clip(z, -50, 50)
    return 1 / (1 + np.exp(-z))

def grad_i(w, x_i, y_i):
    p = sigmoid(x_i @ w)
    return (p - y_i) * x_i
```

```python
def full_gradient(w, X, y):
    n = X.shape[0]
    grad = np.zeros_like(w)
    for i in range(n):
        grad += grad_i(w, X[i], y[i])
    return grad / n
```

```python
def sarah(X, y, lr=0.01, epochs=20, m=None):
    n, d = X.shape
    if m is None:
        m = n

    w = np.zeros(d)
    history = []

    for epoch in range(epochs):
        w_inner = w.copy()

        v = full_gradient(w_inner, X, y)
        w_inner -= lr * v

        for _ in range(1, m):
            i = np.random.randint(0, n)

            grad_current = grad_i(w_inner, X[i], y[i])
            grad_prev = grad_i(w, X[i], y[i])

            v = grad_current - grad_prev + v

            w_inner -= lr * v

            w = w_inner.copy()

        w = w_inner

        history.append(np.linalg.norm(full_gradient(w, X, y)))

    return w, history
```

```python
w_sarah, hist_sarah = sarah(X, y, lr=0.01, epochs=30)
```

```python
plt.plot(hist_sarah)
plt.yscale("log")
plt.xlabel("Epoch")
plt.ylabel("Gradient Norm")
plt.title("SARAH Convergence")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

```python
def generate_data(n=1000, d=10):
    X = np.random.randn(n, d)
    true_w = np.random.randn(d)
    logits = X @ true_w
    y = (logits > 0).astype(int)
    return X, y

X, y = generate_data()
n, d = X.shape
```

```python
def sigmoid(z):
    z = np.clip(z, -50, 50)
    return 1 / (1 + np.exp(-z))

def grad_i(w, x_i, y_i):
    p = sigmoid(x_i @ w)
    return (p - y_i) * x_i
```

```python
def full_gradient(w, X, y):
    n = X.shape[0]
    grad = np.zeros_like(w)
    for i in range(n):
        z = np.clip(X[i] @ w, -50, 50)
        p = 1 / (1 + np.exp(-z))
        grad += (p - y[i]) * X[i]
    return grad / n
```

```python
def adam(X, y, lr=0.01, epochs=20, beta1=0.9, beta2=0.999, eps=1e-8):
    n, d = X.shape
    w = np.zeros(d)
    m = np.zeros(d)
    v = np.zeros(d)

    history = []
    t = 0

    for epoch in range(epochs):
        for _ in range(n):
            t += 1
            i = np.random.randint(0, n)

            g = grad_i(w, X[i], y[i])

            m = beta1 * m + (1 - beta1) * g
            v = beta2 * v + (1 - beta2) * (g ** 2)

            m_hat = m / (1 - beta1 ** t)
            v_hat = v / (1 - beta2 ** t)

            w -= lr * m_hat / (np.sqrt(v_hat) + eps)

        history.append(np.linalg.norm(full_gradient(w, X, y)))

    return w, history
```

```python
def amsgrad(X, y, lr=0.01, epochs=20, beta1=0.9, beta2=0.999, eps=1e-8):
    n, d = X.shape
    w = np.zeros(d)
    m = np.zeros(d)
    v = np.zeros(d)
    v_hat_max = np.zeros(d)

    history = []
    t = 0

    for epoch in range(epochs):
        for _ in range(n):
            t += 1
            i = np.random.randint(0, n)

            g = grad_i(w, X[i], y[i])

            m = beta1 * m + (1 - beta1) * g
            v = beta2 * v + (1 - beta2) * (g ** 2)

            v_hat = v / (1 - beta2 ** t)
            v_hat_max = np.maximum(v_hat_max, v_hat)

            m_hat = m / (1 - beta1 ** t)

            w -= lr * m_hat / (np.sqrt(v_hat_max) + eps)

        history.append(np.linalg.norm(full_gradient(w, X, y)))

    return w, history
```

```python
w_adam, hist_adam = adam(X, y, lr=0.01, epochs=30)
w_amsgrad, hist_amsgrad = amsgrad(X, y, lr=0.01, epochs=30)
```

```python
plt.plot(hist_adam, label="Adam")
plt.plot(hist_amsgrad, label="AMSGrad")
plt.yscale("log")
plt.xlabel("Epoch")
plt.ylabel("Gradient Norm")
plt.title("Adam vs AMSGrad")
plt.legend()
plt.show()
```

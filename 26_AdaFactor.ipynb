```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

```python
def generate_data(n=1000, d=10):
    X = np.random.randn(n, d)
    true_w = np.random.randn(d)
    logits = X @ true_w
    y = (logits > 0).astype(int)
    return X, y

X, y = generate_data()
n, d = X.shape
```

```python
def sigmoid(z):
    z = np.clip(z, -50, 50)
    return 1 / (1 + np.exp(-z))

def grad_i(w, x_i, y_i):
    p = sigmoid(x_i @ w)
    return (p - y_i) * x_i
```

```python
def full_gradient(w, X, y):
    n = X.shape[0]
    grad = np.zeros_like(w)
    for i in range(n):
        z = np.clip(X[i] @ w, -50, 50)
        p = 1 / (1 + np.exp(-z))
        grad += (p - y[i]) * X[i]
    return grad / n
```

```python
def adafactor(X, y, lr=0.01, epochs=20, beta2=0.999, eps=1e-30, clip_threshold=1.0):
    n, d = X.shape
    w = np.zeros(d)

    v = np.zeros(d)

    history = []

    for epoch in range(epochs):
        for _ in range(n):
            i = np.random.randint(0, n)

            g = grad_i(w, X[i], y[i])

            v = beta2 * v + (1 - beta2) * (g ** 2)

            rms = np.sqrt(v + eps)

            update = g / rms

            norm = np.linalg.norm(update)
            if norm > clip_threshold:
                update = update * (clip_threshold / norm)

            w -= lr * update

        history.append(np.linalg.norm(full_gradient(w, X, y)))

    return w, history
```

```python
w_adafactor, hist_adafactor = adafactor(X, y, lr=0.01, epochs=30)
```

```python
plt.plot(hist_adafactor)
plt.yscale("log")
plt.xlabel("Epoch")
plt.ylabel("Gradient Norm")
plt.title("AdaFactor Convergence")
plt.show()
```
